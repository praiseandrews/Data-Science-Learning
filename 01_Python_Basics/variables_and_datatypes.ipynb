```python
# =====================================
# Variables and Data Types in Python
# =====================================

# String
name = "Praisy"

# Integer
age = 23

# Float
height = 5.4

# Boolean
is_student = True

# Display values
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)

print("\n----- Data Types -----")

# Display data types
print("Type of name:", type(name))
print("Type of age:", type(age))
print("Type of height:", type(height))
print("Type of is_student:", type(is_student))

print("\n----- Updating Variables -----")

# Updating a variable
age = 24

print("Updated Age:", age)

print("\n----- More Examples -----")

favorite_food = "Biryani"
favorite_number = 7
price = 99.99
likes_python = True

print("Favorite Food:", favorite_food)
print("Favorite Number:", favorite_number)
print("Price:", price)
print("Likes Python:", likes_python)

print("\n----- Summary -----")

print("int  -> Whole Numbers")
print("float -> Decimal Numbers")
print("str  -> Text")
print("bool -> True or False")
```
