```python
# List

fruits = ["Apple", "Banana", "Mango"]

print("List:", fruits)

# Tuple

colors = ("Red", "Green", "Blue")

print("Tuple:", colors)

# Dictionary

student = {
    "name": "Praisy",
    "age": 23,
    "course": "Data Science"
}

print("Dictionary:", student)

# Set

numbers = {1, 2, 3, 4, 4, 5}

print("Set:", numbers)
```
